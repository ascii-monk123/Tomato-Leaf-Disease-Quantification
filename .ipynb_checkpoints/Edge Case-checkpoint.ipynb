<h1>Custom algorithm to quantify edge cases like the yellow color infection pattern</h1>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
#reading the image
imgNo=41
imgPath='/home/aahan/Documents/Minor Project/Tomato Early Blight/EB{}.jpg'.format(imgNo)

image=cv2.cvtColor(cv2.imread(imgPath),cv2.COLOR_BGR2RGB)

r,g,b=cv2.split(image)

r,g,b=(r.flatten(),g.flatten(),b.flatten())

marker_data=go.Scatter3d(x=r,y=g,z=b,marker=go.scatter3d.Marker(size=3),opacity=0.8,mode='markers')
figure=go.Figure(data=marker_data)
figure.show()
```

<h1> K-means Clustering Start</h1>

```python
#convert from MxNX3 TO KX3 K=MXN and each row is vector in 3d space of rgb i.e r^+g^+b^
shape=image.shape
vector=image.reshape((-1,3))
#convert to float32
vector=np.float32(vector)

#criteria
criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,10,1.0)
K=5
attempts=10
ret,label,center=cv2.kmeans(vector,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)

center=np.uint8(center)
res=center[label.flatten()]
result_img=res.reshape((shape))

plt.imshow(result_img)
```

```python
h_channel=result_img[:,:,0]
plt.imshow(h_channel,cmap='gray')
```

```python
hist,bins,_=plt.hist(h_channel.flatten(),bins=255,range=(0,255),ec='black')
plt.show()
```

```python
values=bins[np.where(hist!=0)]
last=values[len(values)-1]
second=values[2]
print(second)
rows,cols=h_channel.shape
h_dup=[]
for index,ele in enumerate(h_channel.flatten()):
    if ele !=last and ele !=second:
        h_dup.append(0)
    else:
        h_dup.append(255)
h_dup=np.array(h_dup).reshape(rows,cols)
plt.imshow(h_dup,cmap='gray')
```
